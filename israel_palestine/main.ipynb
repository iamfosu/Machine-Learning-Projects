```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fatalities.csv')
df.head(3)
```

Average age of killed ones by citizenship

```python
#average age by countries
average_age = df.groupby('citizenship')['age'].mean()
average_age.plot(kind='bar', figsize=(5,3))
```

Citizenship of who got killed

```python
citizen = df['citizenship'].value_counts()
citizen.plot(kind='bar', figsize=(3,2))
```

Region where they were killed

```python
region = df['event_location_region'].value_counts()
region.plot(kind='bar', figsize=(3,2))
```

Gender of them

```python
gender = df['gender'].value_counts()
gender.plot(kind='bar', figsize=(3,2))
```

Hostilities

```python
hostilities = df['took_part_in_the_hostilities'].value_counts()
hostilities.plot(kind='bar', figsize=(3,2))
```

District of them who were killed

```python
district = df['place_of_residence_district'].value_counts()
district.plot(kind='bar', figsize=(10,5))
```

Types of killing

```python
# type_of_injury = df['type_of_injury'].value_counts()
# type_of_injury.plot(kind='bar', figsize=(5,3))
# type_of_injury.bar_label(type_of_injury.containers[0])
# # for index, value in enumerate(type_of_injury):
# #     plt.text(index, value, str(value), ha='center', va='bottom')

type_of_injury = df['type_of_injury'].value_counts().plot(kind='bar', figsize=(5, 3))
type_of_injury.bar_label(type_of_injury.containers[0]);
```

Ammunition used for killing

```python
ammunition = df['ammunition'].value_counts()
ammunition.plot(kind='bar', figsize=(7,4))
```

Who killed them

```python
killed_by = df['killed_by'].value_counts()
killed_by.plot(kind='bar',figsize=(2,1))
```
